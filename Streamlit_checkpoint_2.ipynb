```python
df.drop(['year','uniqueid'],axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns
```

```python
cat=df.select_dtypes(include='object').columns
for col in cat:
    print(col)
    print(df[col].value_counts())
```

```python
df = df[df['education_level'] != 'Other/Dont know/RTA']
df = df[df['marital_status'] != 'Dont know']
df = df[df['job_type'] != 'Dont Know/Refuse to answer']
```

```python
df.info()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df['bank_account']=df['bank_account'].map({'Yes':1,'No':0})
df['cellphone_access']=df['cellphone_access'].map({'Yes':1,'No':0})
df['location_type']=df['location_type'].map({'Rural':0,'Urban':1})
```

```python
df['gender_of_respondent']=df['gender_of_respondent'].map({'Female':0,'Male':1})
```

```python
le_rel=LabelEncoder()
df['relationship_with_head']=le_rel.fit_transform(df['relationship_with_head'])
```

```python
le_edu=LabelEncoder()
df['education_level']=le_edu.fit_transform(df['education_level'])
```

```python
le_mar=LabelEncoder()
df['marital_status']=le_mar.fit_transform(df['marital_status'])
```

```python
le_job=LabelEncoder()
df['job_type']=le_job.fit_transform(df['job_type'])
```

```python
from sklearn.preprocessing import OneHotEncoder
Ohe=OneHotEncoder(sparse_output=True)
df0=Ohe.fit_transform(df[['country']]).toarray()
```

```python
df0 = pd.DataFrame(df0, columns=Ohe.get_feature_names_out(['country']))
```

```python
df.drop(['index'],axis=1,inplace=True)
```

```python
df.drop(['country'],axis=1,inplace=True)
```

```python
df.info()
```

```python
X=df.drop('bank_account',axis=1)
y=df['bank_account']
```

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
from collections import Counter
```

```python
adasyn = ADASYN(random_state=42)
X_resampled, y_resampled = adasyn.fit_resample(X, y)
print(sorted(Counter(y_resampled).items()))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state= 0)
```

```python
X_train1,X_test1,y_train1,y_test1 = train_test_split(X_resampled, y_resampled ,test_size=0.2,random_state= 0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
param_grid = {
    'criterion': ['gini'],
    'max_depth': [5,10,25],
    'min_samples_split': [2,5,10,15],
    'n_estimators' : [50,80,100]
}
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=0), param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

print("Best parameters found: ", grid_search.best_params_)
print("Best score achieved: ", grid_search.best_score_)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report


predictions = grid_search.best_estimator_.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(accuracy)
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
param_grid1 = {
    'criterion': ['gini'],
    'max_depth': [25,30],
    'min_samples_split': [2,5],
    'n_estimators' : [80,100]
}
grid_search1 = GridSearchCV(estimator=RandomForestClassifier(random_state=0), param_grid=param_grid1, cv=5)
grid_search1.fit(X_train1, y_train1)

print("Best parameters found: ", grid_search1.best_params_)
print("Best score achieved: ", grid_search1.best_score_)
```

```python

from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report


predictions = grid_search1.best_estimator_.predict(X_test1)
accuracy = accuracy_score(y_test1, predictions)
print(accuracy)
from sklearn.metrics import classification_report
print(classification_report(y_test1,predictions))
```
